Vamos a hacer "scraping" a esta página: http://listas.20minutos.es/lista/500-mejores-peliculas-de-la-historia-segun-la-revista-empire-297804/, que contiene un listado de las 500 mejores películas de la historia según la revista Empire.

Su tarea consiste en crear un "Pandas dataframe" que contenga 500 filas (una por película) y las siguientes columnas:
- Ránking (#)
- Título
- Año
- Director
- Reparto
- Sinopsis

Diviértase.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
base_url = "http://listas.20minutos.es/lista/500-mejores-peliculas-de-la-historia-segun-la-revista-empire-297804/"
```

```python
index_html = urlopen(base_url)
index_html
```

```python
sopa = BeautifulSoup(index_html, "lxml")
```

```python
print(sopa.prettify())
```

```python
titles=sopa.find_all("h3")
```

```python
titles=list(titles)
print(titles)
len(titles)
```

```python
titulos=[0]*len(titles)
for i in range(len(titles)):
    titles[i]=str(titles[i])
    ##print(titles[i])
    start=titles[i].find("</strong>")+10
    ##print(start)
    end=titles[i].find("</h3>", start)
    ##print(end)
    titulos[i]=titles[i][start:end]
```

```python
titulos
```

```python
datos=sopa.find_all("div", {"class": "descripcion_elemento_larga"})
```

```python
years=[]

for i in range(len(datos)):
    years.append(re.findall("\d+", datos[i].get_text()))

years 
```

```python
Director=[]

for i in range(len(datos)):
    Director.append(re.findall("Director: (\w+ \w+)", datos[i].get_text()))

Director
```

```python
reparto=[]

for i in range(len(datos)):
    reparto.append(re.findall("Reparto:(\w+ \w.+)", datos[i].get_text()))

reparto
```

```python
sinopsis=[]

for i in range(len(datos)):
    linea=datos[i].get_text()
    numero=re.findall('Sinopsis:[\D+]+', linea)
    sinopsis.append(numero)
    ##sinopsis.append(re.findall("Sinopsis:[\D+ ]+", datos[i].get_text())

sinopsis
```

```python
inicial=sopa.find_all("h3")
ranking=[]
for i in range(len(inicial)):    
    ranking.append(re.findall("\d+", inicial[i].get_text()))
ranking
```

```python
import pandas as pd
df1=pd.DataFrame({"Ranking": ranking})
df2=pd.DataFrame({"Titulo": titulos})
df3=pd.DataFrame({"Año": years})
df4=pd.DataFrame({"Director": Director})
df5=pd.DataFrame({"Reparto": reparto})
df6=pd.DataFrame({"Sinopsis": sinopsis})
```

```python
Datos=pd.concat([df1,df2,df3,df4,df5,df6], ignore_index=False, axis=1)
Datos
```
