### como se puede ver dado que el rango de f(x) es muy pequeño y en cambio el de g es mas amplio por eso al graficarlas f se ve como una linea recta aun cuando intente ajustar el eje Y . ahora voy a intentar graficarlas en dos ejes 

```python
import matplotlib.patches as mpatches
x = np.arange(0, 11);
f = (np.exp(-x/10))*(np.sin(np.pi*x))
g = x*(np.exp(-x/3))
#plt.plot(x, g)
#plt.plot(x, f)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, g, 'g-', label="g(x)= x*e−x/3");
ax2.plot(x, f, 'b-', label="f(x)= e−x/10 * sin(πx)");

ax1.set_xlabel('X data numeros en el intervalo de 0 a 10')
ax1.set_ylabel('g(x)', color='g')
ax2.set_ylabel('f(x)', color='b')
plt.legend(bbox_to_anchor=(1,1.2), loc=0,);
ax1.legend(bbox_to_anchor=(0,1.2), loc=2);

##plt.legend(loc=2,prop={'size':6})


plt.savefig('punto 3.1.jpg')

```

```python
x1=[]
y1=[]
r_1=0.8
for i in range(0,50):
    r = r_1 + np.sin(i)
    x = r*np.cos(i)
    y = r*np.sin(i)
    x1.append(x)
    y1.append(y)
    
x2=[]
y2=[]
r_2=1
for i in range(0,50):
    r = r_2 + np.sin(i)
    x = r*np.cos(i)
    y = r*np.sin(i)
    x2.append(x)
    y2.append(y)
    
x3=[]
y3=[]
r_3=1.2
for i in range(0,50):
    r = r_3 + np.sin(i)
    x = r*np.cos(i)
    y = r*np.sin(i)
    x3.append(x)
    y3.append(y)
```

```python
print(x1[1],x2[1],x3[1])
```

```python
plt.plot(x1,y1, label="limason con r_0 = 0.8");
plt.plot(x2,y2, label="limason con r_0 = 1");
plt.plot(x3,y3, label="limason con r_0 = 1.2");
plt.legend(bbox_to_anchor=(1,1), loc=2,);
plt.xlabel('x = r cos(theta)');
plt.ylabel('y = r sin(theta)');
plt.title(' shape of a limason, given r = r_0 + cos(theta)');

plt.savefig('punto 3.1.pdf')
```
