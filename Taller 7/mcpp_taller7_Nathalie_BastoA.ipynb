Pista: El siguiente código puede serle útil.

```python
# Create function to strip time from date field, and use it to create another column
def to_day(timestamp):
    return timestamp.replace(minute=0,hour=0, second=0)

crimes['Day'] = crimes['Date'].apply(to_day)
```

```python
crimes
```

```python
crimes_by_community_day = crimes.groupby(['Day', 'Community Area'])
crimes_by_community_day_count = crimes_by_community_day['ID'].agg('count')
crimes_by_community_day_count
##d=crimes_by_community_day_count.to_frame()
```

```python
community_crime_timeseries = crimes_by_community_day_count.unstack('Community Area')
community_crime_timeseries
```

```python
community_crime_timeseries[[25,9,73]].plot(); ##la de mas pocos cirmenes,la de mayor numero de crimenes y una intemedia
```
